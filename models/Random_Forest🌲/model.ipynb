##### 1. Import Library

```python
import pandas as pd
import numpy as np
import joblib
import os
```

##### 2. Import Paths ของโมเดลและ scaler

```python
CLASSIFIER_MODEL_PATH = "random_forest_arduino_classifier.pkl"
SCALER_PATH = "arduino_scaler.pkl"
REGRESSOR_MODEL_PATH = "random_forest_vibration_model.pkl"
```

##### 3. โหลดโมเดลและ scaler

```python
rf_cls = joblib.load(CLASSIFIER_MODEL_PATH)
scaler_cls = joblib.load(SCALER_PATH)
rf_reg = joblib.load(REGRESSOR_MODEL_PATH)
```

##### 4. ตัวอย่างข้อมูลใหม่

```python
new_data = pd.read_csv("../../data/processed_data.csv")
```

##### 5. Classification: Predict Health Status

```python
X_new_cls = new_data[['temperature', 'vibration', 'vibration_rms']].copy()
# เติมค่า missing
X_new_cls = X_new_cls.fillna(0)
# Scale features
X_new_cls_scaled = scaler_cls.transform(X_new_cls)
# ทำนาย health status
pred_health = rf_cls.predict(X_new_cls_scaled)
new_data['pred_health_status'] = pred_health

print("Health Status Prediction:")
print(new_data[['temperature', 'vibration', 'vibration_rms', 'pred_health_status']])
```

##### 6. Regression: Predict Vibration

```python
X_new_reg = new_data[['temperature']].copy()
X_new_reg = X_new_reg.fillna(0)
pred_vibration = rf_reg.predict(X_new_reg)
new_data['pred_vibration'] = pred_vibration

print("\nVibration Prediction:")
print(new_data[['temperature', 'vibration', 'pred_vibration']].head(20))
```
