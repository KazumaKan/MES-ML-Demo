##### 1. Import Library

```python
import json
import csv
import pandas as pd
```

##### 2. Convert files

```python
with open("../data/wokwi_output.txt", "r") as f:
    lines = f.readlines()

data = []
for i, line in enumerate(lines, 1):
    line = line.strip()
    if not line:
        continue
    try:
        data.append(json.loads(line))
    except json.JSONDecodeError as e:
        print(f"Skipping line {i} due to JSON error: {e}")

df = pd.DataFrame(data)
df.to_csv("../data/processed_data.csv", index=False)
print("Data saved to ../data/processed_data.csv")
```
