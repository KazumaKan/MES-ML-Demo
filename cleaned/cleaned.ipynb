##### 1. Import Library

```python
import pandas as pd
```

##### 2. อ่านข้อมูล

```python
file_path = '../data/processed_data.csv'
df = pd.read_csv(file_path)
```

##### 3. แปลง timestamp เป็น datetime

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
```

##### 4. ตรวจสอบ missing values

```python
print("Missing values per column:")
print(df.isnull().sum())
```

##### 5. จัดการ missing values

```python
num_cols = ['temperature', 'vibration', 'vibration_rms']
cat_cols = ['machine_id', 'status']

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna('unknown')
```

##### 6. ลบ duplicate ถ้ามี

```python
df.drop_duplicates(inplace=True)
```

##### 7. ตรวจสอบ outlier

```python
for col in num_cols:
    mean = df[col].mean()
    std = df[col].std()
    outliers = df[(df[col] < mean - 3*std) | (df[col] > mean + 3*std)]
    print(f"Outliers in {col}: {len(outliers)} rows")
```

##### 8. บันทึกไฟล์ cleaned

```python
df.to_csv('../data/processed_data_cleaned.csv', index=False)
print("Data cleaning complete. Cleaned file saved.")
```
